```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

def get_info(url):
    # 웹 드라이버 초기화
    driver = webdriver.Chrome(executable_path = 'c:\chromedriver')
    driver.implicitly_wait(3) # or bigger second
    # 열고자 하는 채널 -> 동영상 목록으로 된 url 페이지를 엶
    driver.get(url)
    title_list = list() # 썸네일 제목을 저장하는 리스트
    view_list = list() # 조회수 리스트
    periods_list = list()
    run_time_list = list()
    idx = 1
    while idx < 9: # 다 돌리려면 while == True로 바꾸삼
        try:
            title_xpath = '/html/body/ytd-app/div/ytd-page-manager/ytd-browse/ytd-two-column-browse-results-renderer/div[1]/ytd-section-list-renderer/div[2]/ytd-item-section-renderer/div[3]/ytd-grid-renderer/div[1]/ytd-grid-video-renderer['+str(idx)+']/div[1]/div[1]/div[1]/h3/a'
            viewcnt_xpath = '/html/body/ytd-app/div/ytd-page-manager/ytd-browse/ytd-two-column-browse-results-renderer/div[1]/ytd-section-list-renderer/div[2]/ytd-item-section-renderer/div[3]/ytd-grid-renderer/div[1]/ytd-grid-video-renderer['+str(idx)+']/div[1]/div[1]/div[1]/div[1]/div[1]/div[2]/span[1]'
            period_xpath = '/html/body/ytd-app/div/ytd-page-manager/ytd-browse/ytd-two-column-browse-results-renderer/div[1]/ytd-section-list-renderer/div[2]/ytd-item-section-renderer/div[3]/ytd-grid-renderer/div[1]/ytd-grid-video-renderer['+str(idx)+']/div[1]/div[1]/div[1]/div[1]/div[1]/div[2]/span[2]'
            run_time_xpath = '/html/body/ytd-app/div/ytd-page-manager/ytd-browse/ytd-two-column-browse-results-renderer/div[1]/ytd-section-list-renderer/div[2]/ytd-item-section-renderer/div[3]/ytd-grid-renderer/div[1]/ytd-grid-video-renderer['+str(idx)+']/div[1]/ytd-thumbnail/a/div[1]/ytd-thumbnail-overlay-time-status-renderer/span'
            
            # 한 페이지에 약 8개 불러오는 데, 동영상 목록을 추가 불러오기 위해 스크롤 내림
            if idx % 8 == 0 :
                driver.execute_script('window.scrollBy(0, 1080);')
                time.sleep(2)
                driver.execute_script('window.scrollBy(0, 1080);')
                time.sleep(2)
                driver.execute_script('window.scrollBy(0, 1080);')
                time.sleep(2)


            # 타이틀을 리스트에 저장
            title = driver.find_element_by_xpath(title_xpath)
            title_list.append(title.text)

            # 조회수를 리스트에 저장
            view = driver.find_element_by_xpath(viewcnt_xpath)
            view_list.append(view.text)

            # 시간을 리스트에 저장
            period = driver.find_element_by_xpath(period_xpath)
            periods_list.append(period.text)
            
            
            # 재생시간을 리스트에 저장
            run_time = driver.find_element_by_xpath(run_time_xpath)
            run_time_list.append(run_time.text)
            print(idx, title.text, view.text, period.text, run_time.text)

            idx += 1

        except :
            break
    assert len(view_list) == len(title_list)
    # driver.close()
    return view_list, periods_list, title_list, run_time_list

# 면접왕이형
url1 = 'https://www.youtube.com/c/%EB%A9%B4%EC%A0%91%EC%99%95%EC%9D%B4%ED%98%95/videos'
view1, period1, title1, run_time1 = get_info(url1)
```

```python
title_list = list() # 썸네일 제목을 저장하는 리스트
view_list = list() # 조회수 리스트
date_list = list()
run_time_list = list()

start_url  = 'https://www.youtube.com/c/%EB%A9%B4%EC%A0%91%EC%99%95%EC%9D%B4%ED%98%95/videos'
delay = 3
driver = webdriver.Chrome(executable_path = 'c:\chromedriver')
driver.implicitly_wait(delay)
driver.get(start_url)  
driver.maximize_window()

# 스크롤 높이 가져옴
scroll_height = driver.execute_script("return document.documentElement.scrollHeight")

while True:
    # 스크롤바를 스크롤패인 높이만큼 이동
    driver.execute_script('window.scrollTo(0, document.documentElement.scrollHeight)')

    # 밑에 붙일 내용을 요청해서 화면을 만들때까지 시간차를 둔다.
    time.sleep(2)
    new_scroll_height = driver.execute_script('return document.documentElement.scrollHeight')
    if scroll_height == new_scroll_height:
        # 맨 위로 스크롤
        driver.execute_script('window.scrollTo(0,0)')
        break
    scroll_height = new_scroll_height

        
        
idx = 182

while True:
    
    try:
        driver.execute_script('window.scrollTo(0,0)')
        time.sleep(1)
        

        # x_path 주소 복사할 때 webdriver를 통해 열린 창으로 주소 복사 해야함!! window chrome으로 주소 복사하면 오류!

        # 재생시간을 리스트에 저장
        run_time_xpath = '/html/body/ytd-app/div/ytd-page-manager/ytd-browse/ytd-two-column-browse-results-renderer/div[1]/ytd-section-list-renderer/div[2]/ytd-item-section-renderer/div[3]/ytd-grid-renderer/div[1]/ytd-grid-video-renderer['+str(idx)+']/div[1]/ytd-thumbnail/a/div[1]/ytd-thumbnail-overlay-time-status-renderer/span'
        run_time = driver.find_element_by_xpath(run_time_xpath)
        run_time_list.append(run_time.text)
        time.sleep(1)


        #동영상 클릭
        video = driver.find_element_by_xpath('/html/body/ytd-app/div/ytd-page-manager/ytd-browse/ytd-two-column-browse-results-renderer/div[1]/ytd-section-list-renderer/div[2]/ytd-item-section-renderer/div[3]/ytd-grid-renderer/div[1]/ytd-grid-video-renderer['+str(idx)+']/div[1]/div[1]/div[1]/h3/a').click()
        time.sleep(2)

        # 타이틀을 리스트에 저장
        title_xpath = '//*[@id="container"]/h1/yt-formatted-string'
        title = driver.find_element_by_xpath(title_xpath)
        title_list.append(title.text)
        time.sleep(1)
        print(idx, title.text)
        

        # 조회수를 리스트에 저장
        viewcnt_xpath = '//*[@id="count"]/ytd-video-view-count-renderer/span[1]'
        view = driver.find_element_by_xpath(viewcnt_xpath)
        view_list.append(view.text)

        # 날짜를 리스트에 저장
        date_xpath = '//*[@id="info-strings"]/yt-formatted-string'
        date = driver.find_element_by_xpath(date_xpath)
        date_list.append(date.text)

        driver.back()
        time.sleep(2)

        idx += 1
        
    except:
        break
print(idx, title_list, run_time_list, date_list, view_list)
```
